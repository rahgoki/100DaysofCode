### Day 40!  The topic is JSON

```python
import json
from pprint import pprint
```

```python
with open('mount.json') as f:
    data = json.load(f)
```

```python
for item in data['mounts']:
    pprint(item)
```

---
We can print out each item collected with all of it's attributes like this!

```python
collectedItems = []
for item in data['mounts']['collected']:
    pprint(item)
    collectedItems.append(item)
```

---
Or...We can pull out a specific attribute that we're looking for.  Say we want all of the names for our collected items.  That's easy enough!

```python
for item in data['mounts']['collected']:
    pprint(item['name'])
```

---
We can also create a list of items that meet a criteria.  How about we find all of the aquatic creatures! 

```python
isAquatic = []
for item in data['mounts']['collected']:
    if item['isAquatic'] == True:
        isAquatic.append(item['name'])
```

```python
isAquatic
```

---
This highlights why we should always verify our assumptions.  Quickly scanning the list, I would have guessed that aquatic creatures whould have been 50% of the data.  In reality, it appears that a good portion of our items are aquatic! 

```python
len(isAquatic)
```

```python
len(collectedItems)
```

```python
isNotAquatic = []
for item in data['mounts']['collected']:
    if item['isAquatic'] == False:
        isNotAquatic.append(item['name'])
```

```python
len(isNotAquatic)
```
