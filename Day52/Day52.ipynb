### Day 52!  Feedparser

Feedparser [docs](https://pypi.org/project/feedparser/)

```python
import requests
import feedparser
```

```python
url = 'http://store.steampowered.com/feeds/newreleases.xml'
```

```python
r = requests.get(url)
with open('newreleases.xml', 'wb') as f:
    f.write(r.content)
```

---
When we pull the xml file, the content looks like this!

---
Now let's parse the file and loop over the feed dictionary type object.  

```python
feed_file = 'newreleases.xml'
```

```python
feed = feedparser.parse(feed_file)
```

```python
for entry in feed.entries:
    print(f'{entry.title}: {entry.link}')
```
